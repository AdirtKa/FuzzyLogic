```python
import skfuzzy as fuzz
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display
```

```python
# множество x
x = np.linspace(0, 1, 100)

# три треугольные функции
y1 = fuzz.trimf(x, [0, 0.2, 0.4])
y2 = fuzz.trimf(x, [0.2, 0.5, 0.9])
y3 = fuzz.trimf(x, [0.5, 0.6, 0.9])

# построение графика
plt.figure(figsize=(8, 5))
plt.plot(x, y1, label='y1')
plt.plot(x, y2, label='y2')
plt.plot(x, y3, label='y3')

plt.title("Triangular Member Function Example")
plt.xlabel("x")
plt.ylabel("μ(x)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# множество x
x = np.linspace(0, 1, 100)

# три трапециевидные функции
y1 = fuzz.trapmf(x, [0, 0.2, 0.4, 0.6])
y2 = fuzz.trapmf(x, [0.2, 0.5, 0.6, 0.9])
y3 = fuzz.trapmf(x, [0.5, 0.6, 0.8, 0.9])

# построение графика
plt.figure(figsize=(8, 5))
plt.plot(x, y1, label='y1')
plt.plot(x, y2, label='y2')
plt.plot(x, y3, label='y3')

plt.title("Trapezoidal Member Function Example")
plt.xlabel("x")
plt.ylabel("μ(x)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# множество x
x = np.linspace(0, 1, 100)

# гауссовские функции
y1 = fuzz.gaussmf(x, mean=0.3, sigma=0.1)              # обычная гауссовская
y2 = fuzz.gauss2mf(x, mean1=0.2, sigma1=0.1,
                      mean2=0.4, sigma2=0.2)           # двухпиковая гауссовская

# построение графика
plt.figure(figsize=(8, 5))
plt.plot(x, y1, label='y1')
plt.plot(x, y2, label='y2')

plt.title("Gaussian Member Function Example")
plt.xlabel("x")
plt.ylabel("μ(x)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# множество x
x = np.linspace(0, 1, 100)

# обобщённые колоколообразные функции
y1 = fuzz.gbellmf(x, a=0.6, b=0.5, c=0.4)
y2 = fuzz.gbellmf(x, a=7.2, b=0.2, c=0.2) - 0.25
y3 = fuzz.gbellmf(x, a=5, b=0.35, c=0.8)

# построение графика
plt.figure(figsize=(8, 5))
plt.plot(x, y1, label='y1', color='black')
plt.plot(x, y2, label='y2', color='blue')
plt.plot(x, y3, label='y3', color='green')

plt.title("Generalized Bell Member Function Example")
plt.xlabel("x")
plt.ylabel("μ(x)")

# делаем график «отдалённым»
plt.xlim(-0.1, 1.1)   # немного расширим по X
plt.ylim(0, 1.05)     # теперь видно всю шкалу от 0 до 1

plt.legend()
plt.grid(True)
plt.show()
```

```python
# множество x
x = np.linspace(0, 1, 100)

# одиночные сигмоиды
sig1 = fuzz.sigmf(x, b=0.3, c=12)     # центр 0.3, крутизна 12
sig2 = fuzz.sigmf(x, b=0.7, c=-12)    # центр 0.7, отрицательная крутизна

# произведение (psigmf) = две сигмоиды, задающие "окно"
psig = fuzz.psigmf(x, b1=0.5, c1=15, b2=0.7, c2=-15)

# разность (dsigmf) = "окно" с другими параметрами
dsig = fuzz.dsigmf(x, b1=0.1, c1=15, b2=0.3, c2=15)

# построение графика
plt.figure(figsize=(8, 5))
plt.plot(x, sig1, label='sig1')
plt.plot(x, sig2, label='sig2')
plt.plot(x, psig, label='psig')
plt.plot(x, dsig, label='dsig')

plt.title("Sigmoidal Member Functions Examples")
plt.xlabel("x")
plt.ylabel("μ(x)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# множество x
x = np.linspace(0, 1, 100)

# функции принадлежности
pi = fuzz.pimf(x, a=0.1, b=0.5, c=0.7, d=1.0)   # пи-образная
z  = fuzz.zmf(x, a=0.2, b=0.4)                  # Z-образная
s  = fuzz.smf(x, a=0.6, b=0.8)                  # S-образная

# построение графика
plt.figure(figsize=(8, 5))
plt.plot(x, pi, label='pi')
plt.plot(x, z, label='z')
plt.plot(x, s, label='s')

plt.title("Shaped Member Functions Examples")
plt.xlabel("x")
plt.ylabel("μ(x)")

plt.legend()
plt.grid(True)
plt.show()
```

```python
# множество x
x = np.arange(0, 10.1, 0.1)

# гауссовские функции
y1 = fuzz.gaussmf(x, mean=3, sigma=1.2)
y2 = fuzz.gaussmf(x, mean=7, sigma=1)

# --- операции над функциями ---
# t-norm (AND)
yy1 = np.fmin(y1, y2)         # min
yy3 = y1 * y2                 # dprod (algebraic product)

# s-norm (OR)
yy2 = np.fmax(y1, y2)
display(y1)
display(y2)
display(yy2)
yy2 = np.ones(shape=(len(x), 1))# max
yy4 = y1 + y2 - y1 * y2       # dsum (algebraic sum)

# --- построение графиков ---
plt.figure(figsize=(8, 10))

# MF
plt.subplot(3, 1, 1)
plt.plot(x, y1, label='mf1')
plt.plot(x, y2, label='mf2')
plt.title('Member Function Evaluation')
plt.xlabel('x')
plt.ylabel('μ(x)')
plt.ylim(-0.1, 1.1)
plt.legend()
plt.grid(True)

# AND operations
plt.subplot(3, 1, 2)
plt.plot(x, yy1, label='min')
plt.plot(x, yy3, label='dprod')
plt.title('AND OPERATION')
plt.xlabel('x')
plt.ylabel('and(mf1,mf2)')
plt.ylim(-0.1, 1.1)
plt.legend()
plt.grid(True)

# OR operations
plt.subplot(3, 1, 3)
plt.plot(x, yy2, label='max')
plt.plot(x, yy4, label='dsum')
plt.title('OR OPERATION')
plt.xlabel('x')
plt.ylabel('or(mf1,mf2)')
plt.ylim(-0.1, 1.1)
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
x = np.arange(0, 10.1, 0.1)

# Создаем гауссовскую функцию принадлежности
y1 = fuzz.gaussmf(x, 3, 1.2)  # [центр, ширина]

# Дополнение (NOT операция)
y2 = fuzz.fuzzy_not(y1)  # complement - one's complement

# Создаем график
plt.figure(figsize=(10, 6))
plt.plot(x, y1, label='mf1', linewidth=2)
plt.plot(x, y2, label='Not_mf1', linewidth=2)

plt.title('Member Function and Inverse')
plt.xlabel('x')
plt.ylabel('mu(x)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.xlim(0, 10)
plt.ylim(-0.1, 1.1)
plt.show()
```
